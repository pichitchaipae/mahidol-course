```python
import numpy as np
import pandas as pd

tips = pd.read_csv("tips.csv")
display(tips)
```

```python
tips["tip_pct"] = tips["tip"] / tips["total_bill"]
display(tips)
```

```python
grouped = tips.groupby(["day", "smoker"])
display(grouped["tip_pct"].agg("mean"))
display(grouped["tip_pct"].agg("std"))
```

```python
grouped = tips.groupby(["day", "time", "smoker"])
display(grouped["tip_pct"].agg(["mean", "std", "min"]))
```

```python
def peak_to_peak(arr):
    return arr.max() - arr.min()

display(grouped["tip_pct"].agg(["mean", "std", peak_to_peak]))
```

```python
functions = ["count", "mean", "max"]
result = grouped[["tip_pct", "total_bill"]].agg(functions)
```

```python
display(result["tip_pct"])
```

```python
display(grouped.agg({
        "tip_pct" : ["min", "max", "mean", "std"],
        "size" : "sum"
    })
)
```

```python
grouped_noidx = tips.groupby(["day", "smoker"], as_index=False)

display(grouped_noidx.mean(numeric_only=True))
```

```python
def top(df, n=5, column="tip_pct"):
    return df.sort_values(column, ascending=False)[:n]

display(top(tips, n=6))
```
