### Task 1: Load Titanic Dataset

```python
import numpy as np
import pandas as pd

df = pd.read_csv("C:\\Users\\ASUS\\Downloads\\68-1-itds341-lec8-group-agg\\68-1-itds341-lab8-group-agg\\lab\\titanic-data-clean.csv")
df.head()
```

### Task 2: GroupBy

```python
# Task 2.1: Compute the mean of the Fare of passengers from each Pclass
df_groupby2_1 = df.groupby("Pclass")["Fare"].agg(["mean"])

result2_1 = df_groupby2_1
result2_1
```

```python
# Task 2.2: Compute the standard deviation of the Fare of passengers from each Pclass and Sex
df_groupby2_2 = df.groupby(["Pclass", "Sex"])["Fare"].agg(["std"])
result2_2 = df_groupby2_2
result2_2
```

```python
# Task 2.3: Compute the mean of the Fare of passengers from each Pclass and Age_group
bins = [0, 25, 35, 60, 100]
group_names = ["Youth", "YoungAdult", "MiddleAged", "Senior"]

Age_group = df["Age_group"] = pd.cut(df["Age"], bins, labels=group_names)
df_groupby2_3 = df.groupby(["Pclass", "Age_group"])["Fare"].agg(["mean"])

result2_3 = df_groupby2_3
result2_3
```

```python
# Task 2.4: Create a custom aggregation function
def gt_mean(x): # x is a Fare
    return x[x > x.mean()].count()

df_groupby2_4 = df.groupby(["Embarked", "Sex"])["Fare"].agg([gt_mean])
result2_4 = df_groupby2_4
result2_4
```

### Task 3: Multiple Aggregation Functions

```python
# Task 3.1: Compute the mean and standard deviation of the Fare of passengers from each Pclass and Sex
df_groupby3_1 = df.groupby(["Pclass", "Sex"])["Fare"].agg(["mean", "std"])
result3_1 = df_groupby3_1
result3_1
```

```python
# Task 3.2: Use different aggregation functions on different columns
df_groupby3_2 = df.groupby(["Pclass", "Sex"]).agg({"Fare": ["min", "max", "mean", "std"], "Age": gt_mean, "Parch": "count"})
result3_2 = df_groupby3_2
result3_2
```

### Task 4: Apply on Group

```python
# Task 4: Apply on Group

def top(df, n=3, column="Fare"):
    return df.sort_values(by=column, ascending=False).head(n)

result4 = df.groupby("Pclass", group_keys=True).apply(top)
result4.index = result4.index.set_names(["Pclass", None])

print(result4)
```

### Task 5: Pivot Table

```python
# Task 5.1: Create a pivot table to aggregate the Fare, Parch, and SibSp of passengers in each Pclass group using mean
df_pivot5_1 = df.pivot_table(index="Pclass", values=["Fare", "Parch", "SibSp"], aggfunc="mean")

result5_1 = df_pivot5_1
result5_1
```

```python
# Task 5.2: Extend the code from Task 5.1 to aggregate using count and mean
df_pivot5_2 = df.pivot_table(index="Pclass", values=["Fare", "Parch", "SibSp"], aggfunc=["count", "mean"])

result5_2 = df_pivot5_2
result5_2
```

```python
# Task 5.3: Extend the code from Task 5.1 to further compute the mean of each sex for Fare, Parch, SibSp
df_pivot5_3 = df.pivot_table(index="Pclass", columns="Sex", values=["Fare", "Parch", "SibSp"], aggfunc="mean")

result5_3 = df_pivot5_3
result5_3
```

```python
# Task 5.4: Extend the code from Task 5.3 to also compute the mean from all passengers (use margins=True)
# Hint: you may find the margin argument of the df.pivot_table useful.
df_pivot5_4 = df.pivot_table(index="Pclass", columns="Sex", values=["Fare", "Parch", "SibSp"], aggfunc="mean", margins=True)

result5_4 = df.pivot_table(
    index="Pclass",
    columns="Sex",
    values=["Fare", "Parch", "SibSp"],
    aggfunc="mean",
    margins=True,
    margins_name="All"
)

result5_4
```
